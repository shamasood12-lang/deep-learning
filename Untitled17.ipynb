```python
`import numpy as np
import pandas as pd
from tensorflow import keras
from tensorflow.keras import layers
import matplotlib.pyplot as plt

np.random.seed(42)
```

```python
# Load Fshion-MNIST dataset

(X_train, y_train), (x_test, y_test) = keras.datasets.fashion_mnist.load_data()

# Normalize

X_train = X_train / 255.0
x_test = x_test / 255.0
```

```python
plt.figure(figsize=(6,6))
for i in range(9):
  plt.subplot(3,3,i+1)
  plt.imshow(X_train[i], cmap='gray')
  plt.title(f'Label: {y_train[i]}')
  plt.axis('off')
plt.suptitle("Sample clothing Images")
plt.show()
```

```python
unique, counts = np.unique(y_train, return_counts=True)
plt.figure()
plt.bar(unique, counts)
plt.title("Class distribution")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()
```

```python
model = keras.Sequential([
    layers.Flatten(input_shape=(28,28)),
    layers.Dense(128, activation='relu'),
    layers.Dense(10,activation='softmax')
])
```

```python
from logging import logProcesses
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
history = model.fit(
    X_train , y_train,
    epochs=15,
    validation_data=(x_test, y_test)
)
```

```python
plt.figure()
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.legend()
plt.title("accuracy curve (optimization + generalization)")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.show()
```

```python
plt.figure()
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.legend()
plt.title("loss curve (optimization + generalization)")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.show()
```

```python
test_loss, test_acc = model.evaluate(x_test, y_test)
print("Test Accuracy:",test_acc * 100)
```

```python
preds = model.predict(x_test[:9])

plt.figure(figsize=(6,6))
for i in range(9):
  plt.subplot(3,3,i+1)
  plt.imshow(x_test[i], cmap='gray')
  plt.title(f"P:{np.argmax(preds[i])} A:{y_test[i]}")
  plt.axis('off')
plt.suptitle("Predicions vs Actual")
plt.show()
```
